```python
import configparser
import tweepy
config = configparser.ConfigParser()
config.read('config.ini') ##add your twitter api config file

# Read the config file
api_key = config['twitter']['api_key']
api_key_secret = config['twitter']['api_key_secret']
access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret']

# Authenticate
auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
alhudood_en_user = 'AlHudoodnetEn'
# Set the limit of number of records
limit=270

# Search using pagination
tweets_alhudood_en = tweepy.Cursor(api.user_timeline, screen_name=alhudood_en_user, tweet_mode='extended').items(limit)

# Create columns for the DataFrame
columns = ['Time', 'User', 'Tweet']
alhudood_en_data = []

# Iterate through the results and append them to the list
for tweet in tweets_alhudood_en:
    alhudood_en_data.append([tweet.created_at, tweet.user.screen_name, tweet.full_text])

# Create a dataframe with the results
alhudood_en_df = pd.DataFrame(alhudood_en_data, columns=columns)
```

Preprocessing English Tweets

```python
!pip install emoji
```

```python
from pandas._libs.tslibs import timestamps
from nltk.corpus import stopwords
import nltk
import emoji
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize

def preprocess_tweet_en(tweet):
  tmp = tweet.lower()
  tmp = re.sub('http[^\s]+',' ', tmp) #links
  tmp = re.sub('www[^\s]+',' ', tmp) #links
  tmp = re.sub('@[^\s]+',' ', tmp) #mentions
  tmp = re.sub('#',' ', tmp) #hashtag sign
  #tmp = re.sub(r'[^0-9a-zA-Z]+', ' ', tmp) #keep letters and numbers, remove punc and emojis
  tmp = emoji.replace_emoji(tmp, replace='')
  tmp = re.sub('breaking *(:|-|!)*', ' ', tmp)
  tmp = re.sub('developing *(:|-|!)*', ' ',tmp)
  tmp = re.sub('news *(:|-|!)*', ' ', tmp)
  tmp = re.sub('update *(:|-|!)*', ' ', tmp)
  tmp = tmp.rstrip().lstrip()
  return tmp
```

```python
satire_en_df['tweet'] = satire_en_df['tweet'].map(lambda tweet: preprocess_tweet_en(tweet))
satire_en_df = satire_en_df[satire_en_df['tweet'].str.len() > 35]

real_en_df['tweet'] = real_en_df['tweet'].map(lambda tweet: preprocess_tweet_en(tweet))
real_en_df = real_en_df[real_en_df['tweet'].str.len() > 35]

fake_en_df['tweet'] = fake_en_df['tweet'].map(lambda tweet: preprocess_tweet_en(tweet))
fake_en_df = fake_en_df[fake_en_df['tweet'].str.len() > 35]
```
